```python
# While loop
# Used when iteration are unknown
# Print i as long as i is less than 6

i = 1
while i < 6:
    print(i)
    i = i+1
```

```python
# Print numbers from 1 to N
N = int(input("Enter the value of N: "))
i = 1
while i <= N:
  print(i)
  i = i+1
```

```python
# Factorial of a number
N = int(input("Enter the value of N: "))
i = 1
fact = 1
while i<=N:
  fact = fact*i
  i = i+1
print(fact)
```

```python
# Fibonacci Series
N = int(input("Enter the value of N: "))
b=1
a=0
while N>0:
  c=a+b
  print(a)
  N=N-1
  a=b
  b=c
```

```python
# Reverse a number
N = int(input("Enter the value of N: "))
i = 1
rev = 0
while N>0:
  digit = N%10
  rev = rev*10 + digit
  N= N//10
print(rev)
```

```python
#Palindrome
n = (input("Enter the value of N: "))
i,j = 0,len(n)-1
is_palindrome = True
while i<j:
  if n[i]!=n[j]:
    is_palindrome = False
    break
  i+=1
  j-=1
if is_palindrome:
  print("Yes")
else:
  print("No")
```

```python
#Search an element in list

list = [1,2,3,4,5]
element = int(input("Enter the element to search: "))
n=0
found = False
while list:
  if list[n]==element:
    found = True
    print("Location: ",n)
    break
  else:
    print("element not found")
  n=n+1
```

```python
#While-Continue print numbers from 1-10 skip 5

i=1
while i<11:
  if i==5:
    i+=1
    continue
  print(i)
  i+=1
```

```python
# Multiplication

A=[[1,2,3],[4,5,6],[1,2,1]]
B=[[1,1,1],[6,5,4],[1,2,1]]

lA = len(A)
lB = len(B)
Result = [[0,0,0],[0,0,0],[0,0,0]]
if lA==lB:
  for i in range(3):
    for j in range(3):
      for k in range(3):
        Result[i][j]+=A[i][k]*B[k][j]

  print("Product is")
  for row in range(3):
    for col in range(3):
      print(Result[row][col], end=" ")
    print()
else:
  print("Wrong Order")
```
